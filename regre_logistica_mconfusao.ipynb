```python
import numpy as np
import csv
```

### Leitura do CSV

```python
def load_data(filename): #filename é o caminho do arquivo
  X = []
  y = []

  with open(filename, "r", encoding="utf-8") as f:
    reader = csv.DictReader(f) # Ler o csv e colocar em um dicionário

    for row in reader: # iterando o dict
      #Target
      survived = row['survived']
      if survived == "":
        continue

      y.append(int(survived))

      # Features selecionadas
      # Vamos usar: pclass, sex, age, sibsp, parch, fare

      pclass = float(row["pclass"]) if row["pclass"] != "" else 0

      # Convertendo sexo
      sex = 1 if row["sex"] == "female" else 0

      age = float(row["age"]) if row["age"] != "" else -1

      sibsp = float(row["sibsp"]) if row["sibsp"] != "" else 0
      parch = float(row["parch"]) if row["parch"] != "" else 0
      fare = float(row["fare"]) if row["fare"] != "" else 0

      X.append([pclass, sex, age, sibsp, parch, fare])

  return np.array(X), np.array(y).reshape(-1, 1)
```

### Tratamento de dados

```python
def fill_missing_age(X):
  ages = X[:, 2]
  mean_age = np.mean(ages[ages != -1])

  X[:, 2] = np.where(ages == -1, mean_age, ages)
  return X

# Normalizar (estatística) - para corrigir discrepâncias de escala
def normalize(X):
  mean = np.mean(X, axis=0)
  std = np.std(X, axis=0)
  return (X - mean) / (std + 1e-8)

def add_bias(X):
  ones = np.ones((X.shape[0], 1))
  return np. hstack((ones, X))
```

### Funções do modelo

```python
def sigmoid(z):
  return 1 / (1 + np.exp(-z)) # exp tem o euller encapsulado

def compute_cost(X, y, beta):
  m = len(y)
  h = sigmoid(X @ beta)
  epsilon = 1e-8

  cost = -(1/m) * np.sum(
      y * np.log(h + epsilon) + (1 -y) * np.log(1 - h + epsilon)
  )

  return cost

def gradient_descent(X, y, beta, lr, epochs):
  m = len(y)

  for i in range(epochs):
    h = sigmoid(X @ beta)
    gradient = (1/m) * (X.T @ (h - y))
    beta = beta - lr * gradient

    # if i % 100 == 0:
    #   print(f"Epoch: {i} - Cost: {compute_cost(X, y, beta):.4f}")

  return beta
```

### Treinamento: encontrar os melhores pesos

```python
def train(X, y):
  X = fill_missing_age(X)
  X = normalize(X)
  X = add_bias(X)

  #np.zeros para começar o vetor com pesos zerados - do tamanho de X
  beta = np.zeros((X.shape[1], 1))

  #lr = 0.01 para começar devagar
  beta = gradient_descent(X, y, beta, lr=0.01, epochs=2000)
  return beta, X
```

### Predição e cálculo da Matriz de Confusão

```python
def predict(X, beta):
  probs = sigmoid(X @ beta)
  return (probs >= 0.5).astype(int) # o 0.5 é o decision boundary

def confusion_matrix(y_true, y_pred):
  # Achatar os arrays para garantir que tenham apenas 1 dimensão
  y_t = y_true.flatten()
  y_p = y_pred.flatten()

  # O operador '&' faz a comparação bit a bit (element-wise) nos arrays do numpy
  # Faz AND lógico, elemento por elemento de cada array.
  # Por exemplo:
    # y_t = np.array([1, 0, 1, 1])
    # cond1 = (y_t == 1)
    # Resultado: [True, False, True, True]

  VP = np.sum((y_t == 1) & (y_p == 1)) # Faz a soma das ocorrências VP
  VN = np.sum((y_t == 0) & (y_p == 0)) # Faz a soma das ocorrências VN
  FP = np.sum((y_t == 0) & (y_p == 1)) # Faz a soma das ocorrências FP
  FN = np.sum((y_t == 1) & (y_p == 0)) # Faz a soma das ocorrências FN

  return VP, VN, FP, FN

def calculate_metrics(VP, VN, FP, FN):
  acc = (VP + VN) / (VP + VN + FP + FN)
  prec = VP / (VP + FP)
  rec = VP / (VP + FN)
  spec = VN / (VN + FP)
  f1 = 2 * (prec * rec) / (prec + rec)

  return acc, prec, rec, spec, f1
```

### Split Treino / Teste - dividir o dataset em teste e treino

```python
def train_test_split(X, y, test_size=0.2):
  np.random.seed(42)

  indices = np.arange(len(X))
  np.random.shuffle(indices)

  split = int(len(X) * (1 - test_size))

  train_idx = indices[:split]
  test_idx = indices[split:]

  return X[train_idx], X[test_idx], y[train_idx], y[test_idx]
```
